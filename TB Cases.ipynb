```python
Indicator_list = df.Indicator.unique()
fig, axs = plt.subplots(5, 2, figsize=(15, 25))
fig.suptitle('TB Dashboard in Nigeria', fontsize=14)

for i, Indicator in enumerate(Indicator_list):
    row = i // 2
    col = i % 2
    data = df.loc[df['Indicator'] == Indicator, ['2015', '2016', '2017', '2018', '2019', '2020', '2021', '2022']]
    axs[row, col].plot(data.values[0])
    axs[row, col].set_title(Indicator)
    axs[row, col].set_xlabel('Year')
    axs[row, col].set_ylabel('Cases')
    axs[row, col].set_xticks(range(len(data.columns)))  # Set x-axis ticks
    axs[row, col].set_xticklabels(data.columns, rotation=45)  # Rotate x-axis labels for better readability

plt.tight_layout(rect=[0, 0, 1, 0.96])  # Adjust subplot layout
plt.show()
```

## Apply Linear Regression and Visualization of training and test set

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, mean_absolute_error

def visualize_linear_regression(X, y, variable_name):
    """
    Visualize linear regression results on training and test sets.

    Parameters:
    - X: Input data (independent variable)
    - y: Target data (dependent variable)
    - variable_name: Name of the variable being predicted

    Returns:
    None (plots the results)
    """
    X = np.array(X).reshape(-1, 1)
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=44)

    # Fit the linear regression model
    model = LinearRegression()
    model.fit(X_train, y_train)

    # Print model details
    print('Coefficient of determination(R-Squared):', model.score(X_train, y_train))
    print('Intercept:', model.intercept_)
    print('Slope:', model.coef_)

    # Predictions on both training and test sets
    y_pred_train = model.predict(X_train)
    y_pred_test = model.predict(X_test)

    # Plotting training set
    plt.figure(figsize=(15, 5))

    plt.subplot(1, 2, 1)
    plt.scatter(X_train.flatten(), y_train.tolist(), label='Actual (Training)')  # Use flattened version
    plt.plot(X_train.flatten(), y_pred_train, color='b', label='Predicted (Training)')
    plt.title(f'{variable_name} (Training Set)')
    plt.xlabel('Year')
    plt.ylabel('Cases')
    plt.legend()

    # Plotting test set
    plt.subplot(1, 2, 2)
    plt.scatter(X_test.flatten(), y_test.tolist(), label='Actual (Test)')  # Use flattened version
    plt.plot(X_test.flatten(), y_pred_test, color='r', label='Predicted (Test)')
    plt.title(f'{variable_name} (Test Set)')
    plt.xlabel('Year')
    plt.ylabel('Cases')
    plt.legend()

    plt.tight_layout()
    plt.show()

    # Plotting residuals
    plt.figure(figsize=(15, 5))

    plt.scatter(y_pred_test, y_pred_test - y_test, color='r')
    plt.hlines(y=0, xmin=min(y_test), xmax=max(y_test), color='k', linewidth=2)
    plt.title('Residuals Plot')
    plt.xlabel('Predicted Values')
    plt.ylabel('Residuals')

    plt.show()

    # Additional metrics
    y_pred = model.predict(X_test)

    # Calculate mean squared error
    mse = mean_squared_error(y_test, y_pred)

    # Calculate mean absolute error
    mae = mean_absolute_error(y_test, y_pred)

    # Print the results
    print("Mean Squared Error:", round(mse, 2))
    print("Mean Absolute Error:", round(mae, 2))
    
    print('Previous year(2023) value:',y.loc[7])

    # Prediction for the year 2023
    x_2023 = np.array([2023]).reshape(-1, 1)
    prediction_2023 = model.predict(x_2023)
    print('Prediction for 2023:', prediction_2023)
    
    # Prediction for the year 2024
    x_2024 = np.array([2024]).reshape(-1, 1)
    prediction_2024 = model.predict(x_2024)
    print('Prediction for 2024:', prediction_2024)

# Example usage
# Assuming df['Year'] and df['Cases'] are defined, and variable_name is provided
# visualize_linear_regression(df['Year'], df['TB Case Detection Rate or TB Treatment Coverage (DT-3)'], variable_name='TB Case Detection Rate or Treatment Coverage')
```
